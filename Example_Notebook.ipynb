# Marking from scratch


 Everything is handled currently by the class, so its not that flexible. Below I extracted the mark function so you can apply to any data.
 

```python
from sklearn.gaussian_process.kernels import ConstantKernel, RBF
from sklearn.gaussian_process import GaussianProcessRegressor
from numpy import sin as sin
from numpy import cos as cos

ngrid=256
length_scale=0.75
kernel = 20 * RBF(length_scale=length_scale)

#If using angles....
nodes = Mark_10.nodes #values of mark at nodes
#or...
def convert_from_angle(angles):
        '''convert from 3 angles to 4 points in delta_R space  i.e. cartesian 4D on surface of sphere'''
        a,b,c = angles
                    
        w = sin(a)*sin(b)*sin(c)
        x = sin(a)*sin(b)*cos(c)
        y =sin(a)*cos(b)
        z = cos(a)
        return(np.array([w,x,y,z]))
nodes = convert_from_angle(mark_angles_10)
n_nodes=len(nodes)

delta_R_fid = Mark_10.smoothed_fields['fiducial'] #replace with your smoothed field
delta = Mark_10.fields['fiducial'] #replace with original field

delta_range = [np.min(delta_R_fid), np.max(delta_R_fid)]
delta_R_train = ((np.arange(n_nodes)+0.5)/n_nodes*(delta_range[1]-delta_range[0])+delta_range[0]).reshape(-1,1) #this is the nodes for the GP!
# Use Gaussian Process to create marks
gpr = GaussianProcessRegressor(kernel=kernel, random_state=1)
gpr.fit(delta_R_train, nodes)

mark_field, _ = gpr.predict(delta_R_fid.flatten().reshape(-1, 1), return_std=True)
mark_field = mark_field.reshape([ngrid, ngrid, ngrid])
# Calculate the marked field in real space
marked_field_fid = (delta + 1) * mark_field
marked_field_fid -= np.mean(marked_field_fid)


#then just calculate the power spectrum! :) 
```
